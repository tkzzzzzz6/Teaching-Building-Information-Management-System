```python
# encoding: gbk
import csv
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns

# 设置中文字体和样式
plt.rcParams['font.family'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('seaborn-v0_8-paper')
# seaborn-v0_8-whitegrid or 'seaborn-v0_8-paper'
def count_repetitions_in_column(filename, column_name):
    with open(filename, 'r', encoding='gbk') as csv_file:
        reader = csv.DictReader(csv_file)
        values = [row[column_name] for row in reader]

    counter = Counter(values)
    return counter

# 统计重复次数并可视化
filename = '../data/2024年12月10日上午（第1讲）有课的教室 - 副本.csv'
column_name = '教学楼'

# 读取csv文件并统计重复次数
df = pd.read_csv(filename, encoding='gbk')
result = count_repetitions_in_column(filename, column_name)

# 创建美化后的柱状图
plt.figure(figsize=(12, 7))
bars = plt.bar(result.keys(), result.values(), 
               color=sns.color_palette("pastel", len(result)),  # 使用更柔和的pastel配色
               alpha=0.8, width=0.6)

# 添加数值标签
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

# 设置标题和轴标签
plt.title('教学楼分布统计', fontsize=16, pad=20)
plt.xlabel('教学楼',fontsize=12, labelpad=10)
plt.ylabel('出现次数', fontsize=12, labelpad=10)

# 设置x轴标签旋,使其垂直显示
plt.xticks(range(len(result.keys())), result.keys(), rotation=90)

# 美化坐标轴
plt.grid(True, linestyle='--', alpha=0.7, axis='both')  # 添加横向和纵向网格线，调整透明度

# 添加边框
plt.box(True)

# 调整布局
plt.tight_layout()

plt.show()
```
