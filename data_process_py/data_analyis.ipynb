```python
# encoding: gbk
import csv
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns

# 
plt.rcParams['font.family'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('seaborn-v0_8-paper')
# seaborn-v0_8-whitegrid or 'seaborn-v0_8-paper'
def count_repetitions_in_column(filename, column_name):
    with open(filename, 'r', encoding='gbk') as csv_file:
        reader = csv.DictReader(csv_file)
        values = [row[column_name] for row in reader]

    counter = Counter(values)
    return counter

# 
filename = '../data/202412101 - .csv'
column_name = ''

# csv
df = pd.read_csv(filename, encoding='gbk')
result = count_repetitions_in_column(filename, column_name)

# 
plt.figure(figsize=(12, 7))
bars = plt.bar(result.keys(), result.values(), 
               color=sns.color_palette("pastel", len(result)),  # pastel
               alpha=0.8, width=0.6)

# 
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

# 
plt.title('', fontsize=16, pad=20)
plt.xlabel('',fontsize=12, labelpad=10)
plt.ylabel('', fontsize=12, labelpad=10)

# x,
plt.xticks(range(len(result.keys())), result.keys(), rotation=90)

# 
plt.grid(True, linestyle='--', alpha=0.7, axis='both')  # 

# 
plt.box(True)

# 
plt.tight_layout()

plt.show()
```
